# **<mark>Solenoid Valve</mark>**

```python
# Calculate correlation

correlation = colds[['Solenoid', 'Room_Temp', 'Room_Hum']].corr()

print("Correlation Matrix:")
print(correlation)

# Bifurcate data room-wise
rooms = colds['Room'].unique()

for room in rooms:
    room_data = colds[colds['Room'] == room]
    room_correlation = room_data[['Solenoid', 'Room_Temp', 'Room_Hum']].corr()
    print(f"\nCorrelation Matrix for Room {room}:")
    print(room_correlation)
```

```python
# calculate covariance

covariance_matrix = colds[['Solenoid', 'Room_Temp', 'Room_Hum']].cov()

# Print covariance matrix
print("Covariance matrix between Solenoid, Room_Temp, and Room_Hum:")
print(covariance_matrix)

# Bifurcate changes in temperature and humidity on changes of solenoid room-wise
roomwise_changes = colds.groupby('Room')[['Solenoid', 'Room_Temp', 'Room_Hum']].diff()
```

```python
# Initialize dictionaries to store counts for transitions from 1 to 0 and from 0 to 1
switch_1_to_0 = {}
switch_0_to_1 = {}

# Iterate through each row in the dataset
for index, row in colds.iterrows():
    room = row['Room']
    solenois = row['Solenoid']
    
    # Check if the room exists in the dictionaries
    if room not in switch_1_to_0:
        switch_1_to_0[room] = 0
    if room not in switch_0_to_1:
        switch_0_to_1[room] = 0
    
    # Check if the solenoid switched from 1 to 0
    if index > 0:
        prev_solenois = colds.loc[index - 1, 'Solenoid']
        if prev_solenois == 1 and solenois == 0:
            switch_1_to_0[room] += 1
            
    # Check if the solenoid switched from 0 to 1
    if solenois == 1 and prev_solenois == 0:
        switch_0_to_1[room] += 1

# Print the switch count for each room from 1 to 0
print("Switches from 1 to 0:")
for room, count in switch_1_to_0.items():
    print(f" {room}: {count}")

# Print the switch count for each room from 0 to 1
print("\nSwitches from 0 to 1:")
for room, count in switch_0_to_1.items():
    print(f" {room}: {count}")
```
