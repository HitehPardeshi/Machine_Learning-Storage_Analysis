# **<mark>Solenoid Valve</mark>**

```python
## Program for csv file containing room-solenoid data final ##


# Initialize dictionary to store switches count for each room and date
switch_count = {}

# Iterate through each row in the dataset
for index, row in cold1.iterrows():
    room = row['Room']
    date = row['P_Date'].date()  # Extract date part of the timestamp
    solenoid = row['Solenoid']
    
    # Check if the room exists in the dictionary, if not add it with initial switch count 0
    if room not in switch_count:
        switch_count[room] = {}
    
    # Check if the date exists for the room in the dictionary, if not add it with initial switch count 0
    if date not in switch_count[room]:
        switch_count[room][date] = 0
    
    # Check if the solenois switched from 0 to 1 or 1 to 0
    if index > 0:
        prev_solenoid = cold1.loc[index - 1, 'Solenoid']
        if solenoid != prev_solenoid:
            switch_count[room][date] += 1

# Initialize an empty list to store DataFrames for each room
dfs = []

# Iterate through each date and room to create DataFrames and append them to the list
for room, room_data in switch_count.items():
    room_df = pd.DataFrame.from_dict(room_data, orient='index', columns=[f'{room} '])
    dfs.append(room_df)

# Concatenate all DataFrames along axis 1 (columns) to create the final DataFrame
output_data = pd.concat(dfs, axis=1)

# Fill empty cells with 0
output_data.fillna(0, inplace=True)

# Reset index to make the 'Date' column a regular column instead of an index
output_data.reset_index(inplace=True)
output_data.rename(columns={'index': 'Date'}, inplace=True)

# Write the output DataFrame to a new CSV file
output_data.to_csv('switch_count_may.csv', index=False)
```

```python
# Calculate correlation

correlation = colds[['Solenoid', 'Room_Temp', 'Room_Hum']].corr()

print("Correlation Matrix:")
print(correlation)

# Bifurcate data room-wise
rooms = colds['Room'].unique()

for room in rooms:
    room_data = colds[colds['Room'] == room]
    room_correlation = room_data[['Solenoid', 'Room_Temp', 'Room_Hum']].corr()
    print(f"\nCorrelation Matrix for Room {room}:")
    print(room_correlation)
```

```python
# calculate covariance

covariance_matrix = colds[['Solenoid', 'Room_Temp', 'Room_Hum']].cov()

# Print covariance matrix
print("Covariance matrix between Solenoid, Room_Temp, and Room_Hum:")
print(covariance_matrix)

# Bifurcate changes in temperature and humidity on changes of solenoid room-wise
roomwise_changes = colds.groupby('Room')[['Solenoid', 'Room_Temp', 'Room_Hum']].diff()

# Print room-wise changes
print("\nRoom-wise changes in temperature and humidity on changes of solenoid:")
print(roomwise_changes)
```

```python
target_temp = 4  # Target temperature in Celsius
target_humidity = 50  # Target humidity level in percentage

# Calculate a score for each room based on proximity to target parameters
room_scores = {}
for room, temp_stats in room_temperature.items():
    temp_diff = abs(temp_stats['Mean'] - target_temp)
    humidity_diff = abs(room_humidity[room]['Mean'] - target_humidity)
    score = temp_diff + humidity_diff  # Sum of temperature and humidity differences
    room_scores[room] = score

# Find the room with the minimum score (closest to the target parameters)
ideal_room = min(room_scores, key=room_scores.get)
ideal_score = room_scores[ideal_room]

# Display the room with ideal cold storage parameters
print(f"The room with the ideal cold storage parameters is {ideal_room}.")
print(f"Temperature: {room_temperature[ideal_room]['Mean']:.2f} °C, Humidity: {room_humidity[ideal_room]['Mean']:.2f}%")
print(f"Score: {ideal_score:.2f}")  # Lower score indicates closer proximity to target parameters
```
